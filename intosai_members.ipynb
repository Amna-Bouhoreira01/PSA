```python
with open('full-members.html', "r") as html_file:
    content= html_file.read()
    soup = BeautifulSoup(content,'lxml')
    members = soup.find_all('div', class_='card')
    for member in members:
        country_name= member.span.text
        print(country_name)
```

```python
with open('full-members.html', "r") as html_file:
    content= html_file.read()
    soup = BeautifulSoup(content,'lxml')
    sais_name= soup.find_all('span', itemprop = 'name')
    for sai_name in sais_name:
        sai = sai_name.text
        print(sai)
```
